```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

#### Dimension = 0

```python
nValues = np.array([128,256,512,1024,2048,4096,8192,16384,32768,65536,131072,262144])
averageMSTWeights = [1.19,1.19,1.19,1.19,1.19,1.19,1.19,1.19,1.19,1.19,1.19,1.19]
plt.plot(nValues, averageMSTWeights, 'o', label = 'Data')
plt.plot(nValues, [1.19] * len(nValues), '--', label = '$f(n)=1.19')
plt.xlabel('n')
plt.ylabel('Average MST Weight')
plt.legend()
plt.savefig('Figures/d0.pdf', bbox_inches = 'tight')
```

```python
def powerFunc(x,a,b):
    return a*x**b
```

#### Dimension = 2

```python
nValues = np.array([128,256,512,1024,2048,4096,8192,16384,32768,65536,131072,262144])
averageMSTWeights = [7.60,10.60,14.96,20.99,29.62,41.76,58.94,83.15,117.47,165.99,234.60,331.67]
popt, pcov = curve_fit(powerFunc, nValues, averageMSTWeights)
print(popt)
plt.plot(nValues, averageMSTWeights, 'o', label = 'Data')
plt.plot(nValues, powerFunc(nValues, *popt), '--', label = '$f(n)=$0.66$n^{0.5}$')
plt.xlabel('n')
plt.ylabel('Average MST Weight')
plt.legend()
plt.savefig('Figures/d2.pdf', bbox_inches = 'tight')
```
